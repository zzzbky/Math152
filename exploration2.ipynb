```python
import matplotlib.pyplot as plt
import numpy as np 
import math

#1 random walk two dimensions
xlist = np.zeros(100) # Initialize arrays of zeros.
ylist = np.zeros(100) 
x,y = 0,0 # We start at (0,0).

for t in range(100):  # t will be the "time" parameter for the walk.
    xlist[t] = x
    ylist[t] = y
    degree = np.random.choice([0, 90, 180, 270]) #pick a random direction for movement
    x_step = math.sin(math.radians(degree))
    y_step = math.cos(math.radians(degree))
    x += x_step # update the new x coordinate
    y += y_step # update the new y coordinate
plt.figure(figsize=(5,5))
plt.scatter(xlist, ylist, alpha=0.5, c='gold', s=50)
plt.grid() # Add a grid
plt.xlabel('$x$') 
plt.ylabel('$y$')
plt.title('A random walk in 2 dimensions')
plt.show()


#2 random walk two dimensions with a wall
xlist = np.zeros(100) # Initialize arrays of zeros.
ylist = np.zeros(100) # Initialize arrays of zeros.
x = 0 # We start at (x=0).

for t in range(100):  # t will be the "step" parameter for the walk.
    xlist[t] = x
    ylist[t] = 0
    x_step = np.random.choice([-1,1])
    x += x_step # update the new x coordinate
    if x < 0:
        x = 0 # hits the wall and stays here it was
plt.figure(figsize=(5,5))
plt.scatter(xlist, ylist,alpha=0.5, c='gold', s=50)
plt.grid()
plt.xlabel('$x$')
plt.title('A random walk in one dimension with a wall')
plt.show()

#3 with door ang gravity
xlist = np.zeros(100) # Initialize arrays of zeros.
ylist = np.zeros(100) # Initialize arrays of zeros.
x,y = 0,0 # We start at (0,0).
gravity = 0.1
for t in range(100):  # t will be the "time" parameter for the walk.
    xlist[t] = 0
    ylist[t] = y
    y_step = np.random.choice([-1-gravity,1])
    y += y_step # update the new y coordinate
    if y < 0:
        y = -y # hits the floor and bounces off
plt.figure(figsize=(5,5))
plt.scatter(xlist, ylist,alpha=0.5, c='gold', s=50)
plt.grid()
plt.xlabel('$x$')
plt.title('A random walk in one dimension with a wall')
plt.show()
```
